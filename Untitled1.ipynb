```python
# Filter the DataFrame to include only "Strongly agree" responses in Q20
strongly_agree_df = df[df["Digital Learning is a\nblessing for us"] == "Strongly agree"]

# Count the frequency of "strongly agree" and "strongly disagree" responses in Q5 for "Strongly agree" in Q20
strongly_agree_count = strongly_agree_df["Q5"].value_counts().get("Strongly agree", 0)
strongly_disagree_count = strongly_agree_df["Q5"].value_counts().get("Strongly disagree", 0)

# Calculate the percentage of "strongly agree" and "strongly disagree" responses
total_responses = strongly_agree_count + strongly_disagree_count
pct_strongly_agree = (strongly_agree_count / total_responses) * 100
pct_strongly_disagree = (strongly_disagree_count / total_responses) * 100

# Create a bar plot
labels = ["Strongly agree", "Strongly disagree"]
sizes = [pct_strongly_agree, pct_strongly_disagree]
colors = ['#B8E986', '#B68DF6']

# Set up the bar chart
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(labels, sizes, color=colors, width=0.3, edgecolor='white')

# Customize the chart's appearance

ax.set_xlabel("Q5 Responses", fontsize=12)
ax.set_ylabel("Percentage", fontsize=12)
ax.set_ylim(0, 100)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Display the percentages above each bar
for i, v in enumerate(sizes):
    ax.text(i-0.1, v+1, f"{v:.1f}%", fontsize=12)
plt.gca().set_facecolor('#F0F0F0')
# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame to include only "Strongly agree" responses in Q20
strongly_agree_df = df[df["Digital Learning is a\nblessing for us"] == "Strongly agree"]
strongly_disagree_df = df[df["Digital Learning is a\nblessing for us"] == "Strongly disagree"]

# Group the data by the responses in Q5 and count the frequency
strongly_agree_count = strongly_agree_df["Q5"].value_counts().get("Strongly agree", 0)
strongly_disagree_count = strongly_disagree_df["Q5"].value_counts().get("Strongly disagree", 0)
response_counts = strongly_agree_df["Q5"].value_counts()
response_counts = strongly_disagree_df["Q5"].value_counts()
# Sort the responses in Q5 in ascending order
sorted_responses = sorted(response_counts.index)

# Calculate the percentage of each response category
total_responses = response_counts.sum()
response_percentages = (response_counts / total_responses) * 100

# Create a line chart
plt.plot(sorted_responses, response_percentages, marker='o', color='Blue')

# Customize the chart's appearance
plt.title("Q5 Responses for 'Strongly Agree' in Q20")
plt.xlabel("Q5 Responses")
plt.ylabel("Percentage")
plt.ylim(0, 100)
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Display the plot
plt.show()
```

```python
# Filter the DataFrame to include only "At the time of COVID" responses in Q23
during_COVID_df = df[df["Time of getting dependent\nin elearning Platforms"] == "At the time of Covid-19"]

# Count the frequency of "Yes", "No" and "Unsure" responses in Q18 for "At the time of COVID" responses in Q23
yes_count = during_COVID_df["Improvements in\nacademic performance"].value_counts().get("Yes", 0)
no_count = during_COVID_df["Improvements in\nacademic performance"].value_counts().get("No", 0)
unsure_count = during_COVID_df["Improvements in\nacademic performance"].value_counts().get("Unsure", 0)

# Calculate the percentage of "Yes", "No" and "Unsure" responses
total_responses = yes_count + no_count + unsure_count
pct_yes = (yes_count / total_responses) * 100
pct_no = (no_count / total_responses) * 100
pct_unsure = (unsure_count / total_responses) * 100
# Create a bar plot
labels = ["Yes", "No", "Unsure"]
sizes = [pct_yes, pct_no, pct_unsure]
colors = ['#B8E986', '#B68DF6', '#FF7272']

# Set up the bar chart
fig, ax = plt.subplots(figsize=(6, 4))
ax.bar(labels, sizes, color=colors, width=0.3, edgecolor='white')

# Customize the chart's appearance

ax.set_xlabel("Q18 Responses", fontsize=12)
ax.set_ylabel("Percentage", fontsize=12)
ax.set_ylim(0, 100)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Display the percentages above each bar
for i, v in enumerate(sizes):
    ax.text(i-0.1, v+1, f"{v:.1f}%", fontsize=12)
# Display the plot
plt.show()
```

```python
import seaborn as sns

# Filter the DataFrame to include only "Strongly agree" responses in Q14
strongly_agree_df = df[df["Availability of elearning \nmaterials has increased during COVID-19"] == "Strongly agree"]

# Create a cross-tabulation table between Q14 and Q19
cross_tab = pd.crosstab(strongly_agree_df["Availability of elearning \nmaterials has increased during COVID-19"], 
                        strongly_agree_df["Idea about that topic\nwas cleared perfectly"])

# Generate the heatmap
sns.heatmap(cross_tab, annot=True, fmt="d", cmap="YlGnBu")

# Customize the chart's appearance
plt.xlabel("Q19 Responses", fontsize=12)
plt.ylabel("Q14 Responses", fontsize=12)
plt.title("Heatmap of Q14 and Q19", fontsize=14)

# Display the plot
plt.show()
```
